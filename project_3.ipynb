```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Sales Data.csv")
data
data.shape
```

```python
# HERE WE ARE CHECKING NULL VALUES IN THE DATA SET 
data.isnull() # THEIR IS NO NULL VALUE IN THE DATA SET 
```

```python
data
#Convert Order Date to datetime format
data["Order Date"]=pd.to_datetime(data["Order Date"])
```

```python
#Create new columns like Month, City, and Sales (Quantity * Price)
data["month"]=data["Order Date"].dt.month
data
data["sales"]=data["Quantity Ordered"]*data["Price Each"]
data
```

```python
# sales of each product (Top 10 selling products)
df1=data.groupby(["Product"])["sales"].sum()
df1.sort_values(ascending=False).head(10)
```

```python
# Total monthly sales
df2=data.groupby("Product")["month"].value_counts()
print(df2) # kon se month kon se product ki sale kitni hui ye hai 

df3=data.groupby("month")["sales"].sum()
print(df3) #hr mont sare products ko mil akr kitni sale hui ye hai 

df4=data.groupby("Product")["month"].sum()
print(df4) # total quantity of each product sold out in year
```

```python
# Sales by city
df5=data.groupby('City')['sales'].sum()
print(df5)

# product quantity by city 
df6=data .groupby(["City"])['Quantity Ordered'].sum()
print(df6)
```

```python
# Bar chart: Monthly Sales[df3]
print(df3.sort_values(ascending=False))
plt.figure(figsize=(12,12))
plt.bar(df3.index,df3.values,color='cyan',)
plt.xlabel("months")
plt.grid()
plt.show()
```

```python
# Pie chart: Sales by Category
data
data.groupby(['Product'])['sales'].sum().plot(kind='pie',autopct="%0.1f%%",startangle=90,explode=[0.1,0.1,0.1,0.2,0.2,0.2,0.1,0.1,0.1,0.2,0.2,0.2,0.1,0.1,0.1,0.2,0.2,0.2,0.2])
```

```python
# Line plot: Trend of total sales over time
data.groupby('month')['sales'].sum().plot(kind='line',marker='*',markersize=10,linestyle="dashdot")
plt.grid()
plt.show()
```

```python
# Scatter plot: Price vs Quantity Ordered
# df8=data.groupby('Price Each')['Quantity Ordered'].sum()
plt.figure(figsize=(10,8))
plt.scatter(data['Price Each'],data['Quantity Ordered'],alpha=0.9,color='teal',marker="+")
plt.title("price each vs quantity ordered",size=10)
plt.xlabel("quantity ordered",size=15)
plt.ylabel("price each",size=15)
plt.grid()
plt.tight_layout()
plt.show()
```
