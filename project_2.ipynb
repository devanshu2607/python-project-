```python
# What is the total number of Boxes Shipped for each Sales Person?
df00=data.groupby('Sales Person')['Boxes Shipped'].sum()
print(df00)
plt.plot(df00)
plt.xticks(rotation=70)
```

```python
# What is the average Boxes Shipped per Country?
df01=data.groupby('Country')['Boxes Shipped'].mean()
print(df01)
# Here we are going to use color black which is written in plt plot, plt.plot is a syntax used to create graph where we will use lined style = dash dot,marker = + , marker size = 10 and the size of  the line will be 5 
plt.plot(df01,color='black',linestyle='dashdot',marker='+',markersize=10,linewidth=5) 
plt.xlabel('country')
plt.ylabel('avg')
# Here after implying all the line styles and giving all the colors marker size in. We will label the X axis and Y axis where X axis will be the country and Y axis will be the average
```

```python
# Find the total Amount of sales for each Product.
df02=data.groupby('Product')['Amount']
print(df02.sum())
```

```python
# Find the number of sales transactions for each Country.
df03=data.groupby('Country')['Amount'].sum()
print(df03.sort_values())
```

```python
# What is the maximum Amount earned by each Sales Person?
df04=data.groupby('Sales Person')['Amount'].max()
df05=df04.sort_values(ascending=False)
print(df05)
# Here we use sort values to sort the values in descending order where smallest will come to the topmost and the highest will come to the lowermost that will help us to plot graph easily here we also plot simple normal graph with a red color marker in multiplication(*), marker size 5 ,line grid 3, and line style dash dot with a rotation of 90 using aticks takes and two labels X and Y,X equals to salesperson and Y equals to amount
plt.plot(df05,color='red',marker='*',markersize=5,linewidth=3,linestyle='dashdot')
plt.xticks(rotation=90)
plt.xlabel('sales person name')
plt.ylabel('amount earned max.')
plt.show()
```

```python
# Which Product had the highest total number of Boxes Shipped
df06=data.groupby('Product')['Boxes Shipped'].sum().sort_values(ascending=False).head(1)
df06
```

```python
# 1. Sales Distribution by Product
# Plot a bar chart showing the total Boxes Shipped for each Product.
df08=data.groupby('Product')['Boxes Shipped'].sum()
df08
plt.bar(df08.index,df08.values,color='blue')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
#  Top 5 Sales Persons by Total Sales Amount
# Find the total sales amount per Sales Person and plot a horizontal bar chart.
data['Amount']=data['Amount'].replace('[/$,]', '',regex=True).astype(float)
df09=data.groupby('Sales Person')['Amount'].sum()
df10=df09.sort_values(ascending=False).head(5)
df10

plt.barh(df10.index,df10.values,color='black',linewidth=2)
plt.xlabel("amount")
plt.ylabel("sales person")
plt.grid()

plt.show()
```

```python
# 4. Country-wise Sales Performance
# Create a pie chart showing the percentage of total sales per Country.
data['Amount']=data['Amount'].replace("['/$,]",'',regex=True).astype(float)
df13=data.groupby("Country")["Amount"].sum()
df13
plt.pie(df13,autopct='%0.1f%%')
```

```python
# 5. Sales Distribution by Sales Person
# Plot a box plot to visualize the distribution of Boxes Shipped per Sales Person.
df14=data.groupby('Sales Person')['Boxes Shipped'].sum()
df14.sort_values(ascending=False)

plt.plot(df14,marker='*')
plt.xticks(rotation=90)
plt.show()
```

```python
# 6. Most Sold Product by Country
# Group by Country and Product, find the most sold Product in each Country, and visualize using a bar chart.
df14=data.groupby(['Country','Product'])['Boxes Shipped'].sum().reset_index()
print(df14)
df15=df14.loc[df14.groupby('Country')['Product'].idxmax()]
print(df15)
plt.figure(figsize=(10,5))
plt.bar(df15['Country'],df15['Boxes Shipped'],color='skyblue')
plt.show()
```
